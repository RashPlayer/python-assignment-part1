```python
# task 1 raw data
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
# need variable for storing name of student with 103 roll num.
student_103_name = ""

for student in raw_students:
    # take the raw name
    raw_name = student["name"]

    # using strip() to clean the trailing space and title() for casing
    clean_name = raw_name.strip().title()

    # printing to confirm my logic
    print(f"Cleaned Name: {clean_name}")
```

```python
# need variable for storing name of student with 103 roll num
student_103_name = ""

# creating loop
for student in raw_students:

   # using strip() to clean the trailing space and title() for casing
   # makeing Roll Number a number, instead of text
    clean_name = student["name"].strip().title()
    clean_roll = int(student["roll"])

    # for marks, spliting the single string into a list of strings
    string_marks = student["marks_str"].split(", ")

    # then, creating an empty list and adding the integer versions
    clean_marks = []
    for mark in string_marks:
        clean_marks.append(int(mark))

    # verifying name is valid, assuming default as Valid
    valid_status = "✓ Valid name"

    # Loop through each word in the name
    for word in clean_name.split():
        if not word.isalpha(): # If a word is NOT alphabetical
            valid_status = "✗ Invalid name"

    # collecting name for 103 in loop
    if clean_roll == 103:
        student_103_name = clean_name

    # --- PRINT THE PROFILE CARD ---
    print("================================")
    print(f"Student : {clean_name} {valid_status}")
    print(f"Roll No : {clean_roll}")
    print(f"Marks   : {clean_marks}")
    print("================================\n")


# --- AFTER THE LOOP FINISHES ---
print("--- Final Requirement ---")
print(f"Student 103 (ALL CAPS) : {student_103_name.upper()}")
print(f"Student 103 (lowercase): {student_103_name.lower()}")
```

```python
# Task 2 raw data
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

print(f"--- Grades for {student_name} ---")

# range(len(subjects)) creates a sequence of numbers from 0 to 4.
# This lets us grab the 1st subject and 1st mark, then the 2nd, etc.
for i in range(len(subjects)):
    sub = subjects[i]
    score = marks[i]

    # The conditional logic for grades
    if score >= 90:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "F"

    print(f"{sub}: {score} ({grade})")


# Total and Average
total_marks = sum(marks)
# The round() function for rounding off decimal
avg_marks = round(total_marks / len(marks), 2)

# to get Highest Score
highest_mark = max(marks)
# Find the position of that highest mark in the list
high_index = marks.index(highest_mark)
# Use that position to grab the matching subject
best_subject = subjects[high_index]

# get Lowest Score
lowest_mark = min(marks)
low_index = marks.index(lowest_mark)
worst_subject = subjects[low_index]

print(f"Total marks : {total_marks}")
print(f"Average     : {avg_marks}")
print(f"Highest     : {best_subject} ({highest_mark})")
print(f"Lowest      : {worst_subject} ({lowest_mark})")
```

```python
print("\n--- Enter New Subjects ---")

new_subjects_count = 0

# The loop runs continuously until 'break' is triggered
while True:
    new_sub = input("Enter subject name (or type 'done' to stop): ")

    # Check if the user wants to stop. .lower() ensures "DONE" or "Done" still works.
    if new_sub.lower() == "done":
        break

    # Error handling block
    try:
        new_mark_str = input(f"Enter marks for {new_sub} (0-100): ")
        new_mark = int(new_mark_str) # This will crash if it's not a number, triggering 'except'

        # Check if the number is within the valid range
        if new_mark < 0 or new_mark > 100:
            print("Warning: Marks must be between 0 and 100. Entry skipped.")
        else:
            # Valid input! Add the new data to our original lists
            subjects.append(new_sub)
            marks.append(new_mark)
            new_subjects_count += 1
            print(f"✓ Added {new_sub} with {new_mark} marks.\n")

    except ValueError:
        # If the int() conversion failed, this code runs
        print("Warning: Please enter a valid numerical number. Entry skipped.\n")

# --- AFTER THE WHILE LOOP ENDS ---
print("\n--- Final Update ---")
print(f"New subjects added: {new_subjects_count}")

# Recalculate average using the newly expanded lists
new_total = sum(marks)
new_avg = round(new_total / len(marks), 2)
print(f"Updated Average   : {new_avg}")
```

```python
#task 3 data
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# sett8ng variables with default values
pass_count = 0
fail_count = 0
sum_of_all_averages = 0
topper_name = ""
highest_avg = 0

# Print the table header
# padding the lines by setting :<18 tocreate a column exactly 17 characters wide
print(f"{'Name':<18} | {'Average':<7} | {'Status'}")
#create line
print("-" * 40)

# Main Loop: Grab the name and the list of marks together
for name, marks in class_data:

    # Calculate individual average
    total_marks = sum(marks)
    avg = round(total_marks / len(marks), 2)

    # Add this student's average to our running class total
    sum_of_all_averages = sum_of_all_averages + avg

    # Check if they passed or failed, and update our counts
    if avg >= 60:
        status = "Pass"
        pass_count = pass_count + 1
    else:
        status = "Fail"
        fail_count = fail_count + 1

    # Check if this student is the new highest scorer
    if avg > highest_avg:
        highest_avg = avg     # Update the highest score
        topper_name = name    # Remember their name

    # 3. Print the table row using formatting tricks!
    # :<18 keeps the name column 17 spaces wide
    # :.2f forces the number to show exactly 2 decimal places
    print(f"{name:<18} |  {avg:<5.2f}  | {status}")

# Calculate the final overall class average
class_average = round(sum_of_all_averages / len(class_data), 2)

# Print the final statistics below the table
print("\n--- Class Summary ---")
print(f"Passed          : {pass_count}")
print(f"Failed          : {fail_count}")
print(f"Class Topper    : {topper_name} ({highest_avg})")
print(f"Class Average   : {class_average}")
```

```python
# raw essay data
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# removing trailing spaces
clean_essay = essay.strip()
print("--- Step 1: Cleaned Essay ---")
print(clean_essay)
print("\n")


# adding the Title casing
print("--- Step 2: Title Case ---")
# We just print this. If we saved it as Title Case, the lowercase .count() in Step 3 would fail!
print(clean_essay.title())
print("\n")


# counting Python occurrence for both Python or python
print("--- Step 3: Word Count ---")
python_count = clean_essay.lower().count("python")
print(f"The word 'python' appears {python_count} times.")
print("\n")


# replace Python with Python+emoji
print("--- Step 4: Replaced Text ---")
# Here is your catch! We are overwriting clean_essay so the next steps use the updated text.
clean_essay = clean_essay.replace("python", "Python 🐍")
print(clean_essay)
print("\n")


# splitting into sentences
print("--- Step 5: List of Sentences ---")
# Because we updated clean_essay in Step 4, this list will now contain the emojis
sentences_list = clean_essay.split(". ")
print(sentences_list)
print("\n")


# numbered statements with .
print("--- Step 6: Formatted Sentences ---")
sentence_number = 1

for sentence in sentences_list:
    sentence = sentence.strip()

    # Check if the sentence is missing its period
    if not sentence.endswith("."):
        sentence = sentence + "."

    print(f"{sentence_number}. {sentence}")
    sentence_number = sentence_number + 1
```
